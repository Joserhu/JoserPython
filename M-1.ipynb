```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
import time
from PIL import Image
from IPython.display import clear_output
```

使用隨機函數玩骰子遊戲，玩家先滾動骰子，再換電腦滾動，滾出最高的數字者獲勝。

```python
x = random.randint(1, 6)
y = random.randint(1, 6)

DelayTime = 1



# 隨機滾動骰子數字
for i in range(20):
    k = np.random.randint(1,7)
    clear_output(wait=True)
    #print(k)
    print ("開始滾動骰子-----", k)
    time.sleep(0.1)

 # 顯示玩家骰子
plt.subplot(1,2,1)                       # 將顯示窗口分為一行二列子圖
plt.title('Player')                        #圖片標題
plt.imshow(Image.open(str(x) + '.png')) #顯示骰子圖片
plt.axis('off')                         #不顯示座標尺寸
time.sleep(DelayTime)

# 顯示電腦骰子
plt.subplot(1,2,2)                       # 將顯示窗口分為一行二列子圖
plt.title('Computer')                        #圖片標題
plt.imshow(Image.open(str(y) +'.png'))   # 顯示骰子圖片
plt.axis('off')                         #不顯示座標尺寸
time.sleep(DelayTime)


 # 比較骰子數字大小
   
if x == y:
    print ("平手!")  
elif x > y:
    print ("玩家贏了賽局，電腦輸了!")            
elif x < y:
    print ("電腦贏了賽局，玩家輸了!")
```
